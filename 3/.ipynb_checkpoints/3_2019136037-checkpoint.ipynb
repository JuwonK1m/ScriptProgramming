# 문제 1번
### 내용
팬을 표현하는 Fan 클래스를 구현하시오.
### 코드 설명
Fan 클래스의 제일 위에 클래스 변수인 SLOW, MEDIUM, FAST를 선언 및 초기화하였다.  
생성자에서는 각각의 private 인스턴스 변수(speed, on, radius, color)들을 초기화하였다. 매개변수에 기본값을 넣어놓았고, 변수명 앞에 언더바 2개를 붙여서 private 변수로 설정하였다.  
그리고 각각의 인스턴스 변수들에 대한 설정자와 접근자를 만들었다.   
특히 setSpeed() 메소드에서는 인자로 들어온 speed의 값이 클래스 변수의 SLOW(1), MEDIUM(2), FAST(3) 중 어느 하나에도 포함되지 않으면 초기화가 되지 않게 하였고, setRadius() 메소드에서는 인자로 들어온 radius의 값이 음수라면 초기화 되지 않게 하였다.  
  
테스트에서는 Fan 클래스에 대한 생성자 호출을 통해 객체 2개를 생성하여 접근자를 이용해 각각의 인스턴스 변수값을 출력하였다. 

```python
class Fan:
    SLOW = 1
    MEDIUM = 2
    FAST = 3
    
    def __init__(self, speed = SLOW, on = False, radius = 5, color = "blue"):
        self.__speed = speed
        self.__on = on
        self.__radius = radius
        self.__color = color
        
    def setSpeed(self, speed):
        # 매개변수 speed의 값이 SLOW(1) 또는 MEDIUM(2) 또는 FAST(3)인 경우에만 적용
        if (speed == self.SLOW) or (speed == self.MEDIUM) or (speed == self.FAST):
            self.__speed = speed
        
    def getSpeed(self):
        return self.__speed
    
    def setOn(self, on):
        self.__on = on
        
    def getOn(self):
        return self.__on
    
    def setRadius(self, radius):
        if radius >= 0:
            self.__radius = radius
        
    def getRadius(self):
        return self.__radius
    
    def setColor(self, color):
        self.__color = color
        
    def getColor(self):
        return self.__color
    
fan1 = Fan(Fan.FAST, True, 10, "yellow")
fan2 = Fan(Fan.MEDIUM) # 기본값: False, 5, "blue"

print("<fan1 객체>")
print("속도: ", fan1.getSpeed())
print("전원: ", fan1.getOn())
print("반지름: ", fan1.getRadius())
print("색상: ", fan1.getColor())
print("<fan2 객체>")
print("속도: ", fan2.getSpeed())
print("전원: ", fan2.getOn())
print("반지름: ", fan2.getRadius())
print("색상: ", fan2.getColor())
```

### 실행 결과

색상:  yellow  
<fan2 객체>  
속도:  2  
전원:  False  
반지름:  5  
색상:  blue

# 문제 2번
### 내용
Elevator 클래스를 구현하시오.
### 코드 설명
생성자에서는 각각의 인스턴스 변수들을 초기화하였다.  
set_people() 메소드에서는 people이 0 미만이면 초기화할 수 없게 하였고, set_floor() 메소드에서는 floor가 -2 미만이거나 15 초과이면 초기화할 수 없게 하였다.  
10명 이상이 되면 무게 초과이기 때문에, people_in() 메소드에서 people을 1 추가시킨후, 그 값이 10 이상이라면 is_ovarload 변수를 True로 바줘주었다.  
테스트의 무한 반복문에서는 제일 처음에 엘리베이터가 무게 초과(is_overload가 True)이면 반복문을 종료시켰고, 아니라면 다음 로직으로 넘어가 입력 값에 대한 처리들을 하였다.  
입력 값이 "in"일때는 people_in(), "out"일때는 people_out()을 호출하게 하였고, "off"라면 반복문을 종료시켰다.
"in", "out", "off" 입력을 제외한 나머지 입력들은 전부 숫자이기 때문에, int타입으로 변환하여 처리하였다.

```python
class Elevator:
    def __init__(self):
        self.__people = 0
        self.__floor = 1
        self.__is_overload = False
        self.__is_on = False
        
    def set_people(self, people):
        if people >= 0:
            self.__people = people
    
    def get_people(self):
        return self.__people
    
    def set_floor(self, floor):
        if floor >= -2 and floor <= 15:
            self.__floor = floor
            
    def get_floor(self):
        return self.__floor
    
    def set_is_overload(self, is_overload):
        self.__is_overload = is_overload
        
    def get_is_overload(self):
        return self.__is_overload
    
    def set_is_on(self, is_on):
        self.__is_on = is_on
        
    def get_is_on(self):
        return self.__is_on
    
    def people_in(self):
        self.__people += 1
        
        if self.__people >= 10:
            self.__is_overload = True
        
    def people_out(self):
        if self.__people >= 1:
            self.__people -= 1
            
    def power_off(self):
        self.__is_on = False
        
    def print_elevator(self):
        print("현재 사람수:", self.__people)
        print("현재 층:", self.__floor)
        
elevator = Elevator()
elevator.set_is_on(True)
while True:
    if elevator.get_is_overload():
        elevator.power_off()
        print("엘리베이터가 종료됩니다.")
        break
    
    elevator.print_elevator()
    i = input("입력하시오: ")
    print()
    
    if i == "in":
        elevator.people_in()
    elif i == "out":
        elevator.people_out()
    elif i == "off":
        elevator.power_off()
        print("엘리베이터가 종료됩니다.")
        break
    else:
        floor = int(i)
        elevator.set_floor(floor)
```

# 문제 3번
### 내용
Time이라는 이름의 클래스를 구현하시오.
### 코드 설명
생성자에서는 time 모듈의 gmtime() 함수를 사용하였다. time() 함수를 사용할수도 있지만, gmtime()은 멤버 변수별로 시간, 분, 초 등에 대한 데이터를 가지고 있어 편리하다는 장점이 있어 사용하였다. 각각 tm_hour, tm_min, tm_sec 값을 가져와서 time 클래스의 인스턴스 변수에 초기화하였다.  
시간은 0시부터 23시까지 있고, 분은 0분부터 59분까지 있으며, 초는 0초부터 59초까지 있으므로 설정자를 정의할때 그에 따라 조건을 검사하였다.
setTime() 메소드에서는 인자로 들어온 시간을 통해 시간, 분, 초를 계산하여 대입하는 과정을 거쳤다.
인자로 들어온 시간을 x라고 하면 다음과 같다.
- hour는 24시간으로 나눈 나머지를 1시간으로 나눈 몫이므로 **x % (3600 * 24) / 3600**
- minute는 24시간으로 나눈 나머지를 1시간으로 나눈 나머지에서 1분으로 나눈 몫이므로 **x % (3600 / 24) % 3600 / 60**
- second는 24시간으로 나눈 나머지를 1시간을 나눈 나머지에서 1분으로 나눈 나머지이므로 **x % (3600 / 24) % 3600 % 60**

```python
from time import gmtime

class Time:
    def __init__(self):
        # UTC 시간
        now = gmtime()
        self.__hour = now.tm_hour
        self.__minute = now.tm_min
        self.__second = now.tm_sec
        
    def setHour(self, hour):
        if hour >= 0 and hour < 24:
            self.__hour = hour
            
    def getHour(self):
        return self.__hour
    
    def setMinute(self, minute):
        if minute >= 0 and minute < 60:
            self.__minute = minute
            
    def getMinute(self):
        return self.__minute
    
    def setSecond(self, second):
        if second >= 0 and second < 60:
            self.__second = second
            
    def getSecond(self):
        return self.__second
    
    def setTime(self, elapseTime):
        remainder = elapseTime % 86400
        hour = int(remainder / 3600)
        minute = int(remainder % 3600 / 60)
        second = int(remainder % 3600 % 60)
        
        self.__hour = hour
        self.__minute = minute
        self.__second = second
        
time = Time()
print("현재 시간은", "{}:{}:{}".format(time.getHour(), time.getMinute(), time.getSecond()), "입니다.")
elapseTime = eval(input("소요 시간을 입력하세요: "))
time.setTime(elapseTime)
print("현재 시간은", "{}:{}:{}".format(time.getHour(), time.getMinute(), time.getSecond()), "입니다.")
```

### 실행 결과
현재 시간은 10:37:33 입니다.  
소요 시간을 입력하세요: 55550505  
현재 시간은 22:41:45 입니다.

# 문제 4번
### 내용
홍길동 씨의 주민등록번호는 951105-1111111이다. 홍길동씨의 주민등록번호를 연월일(YYYYMMDD)부분과 그 뒤의 숫자 부분으로 나누어 출력하는 프로그램을 작성하시오.
### 코드 설명
주민등록번호를 문자열로 저장하고, 하이픈(-)이 있는 인덱스를 find() 메소드를 통해 찾는다.  
하이픈이 있는 인덱스를 hyphenIndex라고 하고, 슬라이싱 연산자를 이용하여 0부터 (hyphenIndex - 1)까지의 내용과 hyphenIndex부터 마지막 인덱스까지의 내용을 각각 출력하였다.

```python
birthDate = "951105-1111111"

# 하이픈(-)이 있는 인덱스를 찾는다.
hyphenIndex = birthDate.find('-')

print("생년월일:", birthDate[0 : hyphenIndex])
print("뒷자리:", birthDate[hyphenIndex + 1 : len(birthDate)])
```

### 실행 결과
생년월일: 951105  
뒷자리: 1111111
