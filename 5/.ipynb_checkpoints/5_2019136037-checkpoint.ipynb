# 문제 1번
### 내용
n * n 2차원 리스트를 선언합니다. n은 3~10 사이의 값으로 사용자에게 입력 받습니다. 리스트의 원소는 0 또는 1로 랜덤하게 초기화합니다. 리스트의 모든 원소를 출력한 후 리스트 내에서 0인 값을 -1로 바꿔주고 이를 다시 출력합니다.
### 코드 설명
리스트의 원소를 0 또는 1로 랜덤하게 초기화하기위해 random 모듈의 randint() 메소드를 import하였다.
matrix라는 이름의 리스트를 선언하고 횟수 n만큼 반복문을 돌면서 append([])를 호출해주어 2차원 리스트를 만들며, 해당 위치 리스트에서 append(randint(0, 1))를 호출해주어 랜덤한 값으로 전부 초기화한다.
그리고 해당 2차원 리스트를 한번 출력하고, 그 다음에는 이중 반복문을 통해 값이 0인 부분을 -1로 바꾸어주어 다시 한번 출력하였다.

```python
from random import randint

def printMatrix(matrix):
    for i in range(0, len(matrix)):
        for j in range(0, len(matrix[i])):
            print("%3d"%(matrix[i][j]), end = '')
        print()

matrix = []

n = eval(input("n을 입력하시오: "))

for i in range(0, n):
    matrix.append([])
    for j in range(0, n):
        matrix[i].append(randint(0, 1))
        
printMatrix(matrix)
print()
    
for i in range(0, len(matrix)):
    for j in range(0, len(matrix[i])):
        if matrix[i][j] == 0:
            matrix[i][j] = -1
            
printMatrix(matrix)
```

# 문제 2번
### 내용
n * n 2차원 리스트를 생성합니다. n은 3~10 사이의 값으로 사용자에게 입력 받습니다. 리스트가 다음과 같이 되도록 만드시오.

실행 예시 n = 3인 경우  
1  2  3  
6  5  4  
7  8  9  

실행 예시 n = 4인 경우  
1  2  3  4  
8  7  6  5  
9  10 11 12  
16 15 14 13  

위와 같이 리스트가 출력되어야 합니다.  

### 코드 설명
기본적으로는 이중반복문을 통하여 각 행마다 데이터를 추가하였다. (문제 1번과 같이 append() 메소드 사용)  
하지만 숫자의 배치가 아래로 내려가면서 지그재그 형태를 만들기 위하여 다음과 같은 과정의 코드를 작성하였다.  
- 행이 0을 포함한 짝수일때는(row % 2 == 0), 값을 1씩 높여가면서 append() 메소드로 총 n번 값을 추가
- 행이 홀수일때는(row % 2 != 0), 일단 append() 메소드로 총 n번 None을 추가한 다음, 끝 인덱스부터 처음 인덱스로 되돌아오면서 값을 1씩 높여가며 해당 위치에 값을 대입

```python
def printMatrix(matrix):
    for i in range(0, len(matrix)):
        for j in range(0, len(matrix[i])):
            print("%-3d"%(matrix[i][j]), end = '')
        print()

val = 1
matrix = []

n = eval(input("n을 입력하시오: "))

for i in range(0, n):
    matrix.append([])
    if i % 2 == 0:
        for j in range(0, n):
            matrix[i].append(val)
            val += 1
    else:
        for j in range(0, n):
            matrix[i].append(None)
        for j in range(n - 1, -1, -1):
            matrix[i][j] = val
            val += 1
        
printMatrix(matrix)
```

# 문제 3번
### 내용
두 개의 10 * 10 크기의 2차원 리스트를 선언합니다. 리스트 초기 값은 1~9 사이의 정수 값으로 랜덤하게 초기화 합니다. 두 개의 리스트를 A, B라고 했을 때, 두 리스트의 각 원소를 더합니다. 더한 값이 10 미만이면 그대로, 더한 값이 10 이상이면 10을 빼고 이를 출력합니다.
### 코드 설명
문제 1번과 같은 방식으로 2차원 리스트를 총 3개를 만든다. 
해당 2차원 리스트 3개를 각각 matrix1, matrix2, matrix3 라고 하면  
matrix1과 matrix2를 이중반복문을 돌면서 random모듈의 randint() 메소드로 1~9 중의 랜덤한 값을 추가하고, 리스트3에는 None을 추가한다.
그리고 한번 더 이중반복문을 돌면서 matrix1의 값과 matrix2의 값을 더하여 matrix3에 저장하고, 만약 그 값이 10보다 크다면 10을 빼서 다시 저장하였다.
그리고 각각의 2차원 리스트들을 matrix1, '+', matrix2, '=', matrix3 순으로 출력하였다.

```python
from random import randint

def printMatrix(matrix):
    for i in range(0, len(matrix)):
        for j in range(0, len(matrix[i])):
            print("%-3d"%(matrix[i][j]), end = '')
        print()

matrix1 = []
matrix2 = []
matrix3 = []

n = eval(input("n을 입력하시오: "))

for i in range(0, n):
    matrix1.append([])
    matrix2.append([])
    matrix3.append([])
    for j in range(0, n):
        matrix1[i].append(randint(1, 9))
        matrix2[i].append(randint(1, 9))
        matrix3[i].append(None)

for i in range(0, n):
    for j in range(0, n):
        matrix3[i][j] = matrix1[i][j] + matrix2[i][j]
        if matrix3[i][j] >= 10:
            matrix3[i][j] -= 10
        
printMatrix(matrix1)
print('+')
printMatrix(matrix2)
print('=')
printMatrix(matrix3)
```

### 실행 결과
'+' 문자가 마크다운 요소로 인식되어 따로 작성하지 않았습니다. 직접 실행해보는 결과로 대체합니다.

# 문제 4번
### 내용
test_1.txt 파일 중 savage라는 단어가 몇 번 등장하는지 출력하고, savage를 Savage로 변경하여 파일 내용을 출력하는 프로그램을 작성하시오.
### 코드 설명
파일의 내용을 open() 함수를 통해 읽어와 객체를 file 변수에 저장한다. encoding='UTF8' 설정까지 넣어주었는데, 한글 처리를 위해서 넣어준 설정이다. 그리고 file변수에서 readlines() 메소드 호출을 통해 lines변수에 줄의 리스트를 저장하였다.  
그 다음 for문을 통해 한줄씩 읽으면서, 해당 줄에서 split() 메소드를 호출하여 줄별로 단어 리스트를 읽어와 해당 단어 리스트를 순회하면서 savage를 Savage로 바꾸어 출력하고, 카운트 변수를 1 높인다.  
그리고 마지막에는 카운트 변수를 이용하여 savage가 나온 횟수를 출력하였다.  
그리고 file변수에서 close()를 호출하여 파일 객체를 닫아주었다.

```python
file = open("test_1.txt", "r", encoding='UTF8')

lines = file.readlines()

cnt = 0

for line in lines:
    words = line.split()
    for word in words:
        if word == "savage":
            print("Savage", end = ' ')
            cnt += 1
        else:
            print(word, end = ' ')
    print()
    
print()
print("savage는 {}번 등장합니다.".format(cnt))

file.close()
```

# 문제 5번
### 내용
test_1.txt 파일을 읽어서 각 줄마다 번호를 출력하고 2칸 띄워서 원래 텍스트를 출력하는 프로그램을 작성하시오. 
### 코드 설명
open() 함수를 이용해 파일을 읽어오고, readlines() 메소드를 통해 줄의 리스트를 가져온다.    
그리고 for문을 통해 한줄씩 순회하면서, 줄의 번호와 줄의 내용을 형식에 맞게 출력하였다.  
그리고 close() 함수를 호출하여 객체를 닫아주었다.

```python
file = open("test_1.txt", "r", encoding='UTF8')

lines = file.readlines()

for i in range(0, len(lines)):
    print("{}  {}".format(i + 1, lines[i]))
    
file.close()
```
